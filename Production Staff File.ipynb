```python
def split_name(Name):
    splitval = str(Name).split(',')
    last_name, first_name = splitval[0], splitval[1]
    return [last_name, first_name]
```

```python
df2['EmployeeLastName'], df2['EmployeeFirstName'] = zip(*df2.apply(lambda x: split_name(x['EmployeeName']), axis=1))
```
